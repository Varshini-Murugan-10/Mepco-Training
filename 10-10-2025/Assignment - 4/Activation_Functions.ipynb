```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))   # stability trick
    return exp_x / np.sum(exp_x)
```

```python
z = np.linspace(-10, 10, 100)
#Activation function
sigmoid_viz = sigmoid(z)
tanh_viz = tanh(z)
relu_viz = relu(z)
leaky_relu_viz = leaky_relu(z)
softmax_viz = softmax(z)
# Plot
plt.figure(figsize=(12, 8))
plt.plot(z, sigmoid_viz, label='Sigmoid')
plt.plot(z, tanh_viz, label='Tanh')
plt.plot(z, relu_viz, label='ReLU')
plt.plot(z, leaky_relu_viz, label='Leaky ReLU')
plt.plot(z, softmax_viz, label='softmax')
plt.title('Common Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```
