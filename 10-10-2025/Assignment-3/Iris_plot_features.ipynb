```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.figure(figure=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,color='blue',)
```

```python
data = sns.load_dataset("iris")
plt.figure(figsize=(6,4))
sns.scatterplot(data=data,x="sepal_length",y="sepal_width",hue="species",palette="rocket")
plt.title("Iris Dataset with pastel colour palette")
plt.show()
```
