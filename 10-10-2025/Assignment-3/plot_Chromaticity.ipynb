```python
import numpy as np
import imageio.v2 as imageio
import colour
from colour.plotting import plot_RGB_chromaticities_in_chromaticity_diagram_CIE1931

def plot_image_chromaticity(image_path):
        image = imageio.imread(image_path)
        image_float = image.astype(np.float32) / 255.0
        image_linear = colour.models.eotf_inverse_sRGB(image_float)
        pixels_rgb = image_linear.reshape(-1, 3)
        figure, axes = plot_RGB_chromaticities_in_chromaticity_diagram_CIE1931(
            pixels_rgb,
            colourspace='sRGB',
            show_diagram_colours=True,
            show_spectral_locus=True,
            show_whitepoints=True,
            title=f'Chromaticity of Pixels in {image_path}',
            scatter_kwargs={'s': 2, 'alpha': 0.5, 'c': pixels_rgb}
        )
        import matplotlib.pyplot as plt
        plt.show()

image = '/content/peppers.png'
plot_image_chromaticity(image)
```
