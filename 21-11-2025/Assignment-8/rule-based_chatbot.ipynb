***Simple Rule‑Based Chatbot for FAQs***

```python
faq_dataset = {
    "what is your name": "I am a simple rule-based FAQ chatbot.",
    "how are you": "I'm functioning as expected!",
    "what is ai": "AI stands for Artificial Intelligence — machines performing tasks that usually require human intelligence.",
    "what is machine learning": "Machine Learning is a subset of AI where systems learn patterns from data.",
    "bye": "Goodbye! Have a great day!"
}
faq_dataset
```

```python
def chatbot(user_input):
    user_input = user_input.lower().strip()
    for question, answer in faq_dataset.items():
        if question in user_input:
            return answer
    return "Sorry, I don't understand that. Please try asking something else."

# Test
chatbot("What is AI?")
```

```python
user_query = "How are you?"
chatbot(user_query)
```
