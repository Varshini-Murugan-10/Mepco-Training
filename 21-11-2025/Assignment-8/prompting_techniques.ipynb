```python
faq_dataset = {
    "what is python": "Python is a high-level programming language.",
    "what is ai": "AI stands for Artificial Intelligence.",
    "what is machine learning": "Machine learning is a subset of AI focused on pattern learning."
}
faq_dataset
```

```python
def rule_based_chatbot(query):
    q = query.lower().strip()
    for key, ans in faq_dataset.items():
        if key in q:
            return ans
    return "No exact match found."

rule_based_chatbot("What is AI?")
```

***Zero-shot Prompting***

```python
query = "Explain Python."
response = f"Zero-shot response: {rule_based_chatbot(query)}"
response
```

***One-shot Prompting***

```python
example = "Example: What is Python? -> Python is a high-level programming language."
query = "Explain Python."
response = f"Using one-shot example:\n{example}\nOutput: {rule_based_chatbot('what is python')}"
response
```

***Few-shot Prompting***

```python
examples = [
    "What is Python? -> Python is a high-level programming language.",
    "What is AI? -> AI stands for Artificial Intelligence."
]

query = "Explain Machine Learning."
response = "Few-shot examples:\n" + "\n".join(examples) + "\nOutput: " + rule_based_chatbot("what is machine learning")
response
```

***Chain-of-Thought Prompting***

```python
query = "Explain AI in steps."

steps = [
    "Step 1: Identify keywords → AI",
    "Step 2: Match dataset entry",
    "Step 3: Retrieve definition"
]

final_answer = rule_based_chatbot("what is ai")

"\n".join(steps) + "\nFinal Answer: " + final_answer
```
