***Extractive summary***

```python
text="Artificial intelligence (AI) is the ability of machines to perform tasks that typically require human intelligence, such as learning from experience, understanding language, making decisions, and recognizing patterns. AI systems work by using data to simulate human reasoning, and they can improve their accuracy over time. AI is used in many areas, from virtual assistants and recommendation engines to healthcare diagnostics, finance, and autonomous vehicles, and it can automate repetitive tasks and provide valuable insights."
from summarizer import Summarizer
bert_model=Summarizer()
bert_summary=bert_model(text)
bert_summary
```

***Abstractive summary***

```python
from transformers import pipeline
abstractive_summarizer=pipeline("summarization",model="facebook/bart-large-cnn")
abs_sum=abstractive_summarizer(text, max_length=60, min_length=30, do_sample=False)
print(abs_sum[0])
```

***QA gen***

```python
from transformers import pipeline
qa_model=pipeline("text2text-generation",model="t5-base")
context={"The Himalayas are the world's highest mountain range, formed by the collision of the Indian and Eurasian tectonic plates.",
         "The ranges form an arc running for about 2,400 km and varying in width from 150 to 400 km.",
         "The Himalayas are the source of three major Asian river systems: the Indus, Ganges, and Brahmaputra."}
question="What is the world's highest mountain range?"
input_text = f"question: {question} context: {' '.join(context)}"
res=qa_model(input_text)
print(res[0])
```

```python
question="Does any river originate from himalayas?"
input_text = f"question: {question} context: {' '.join(context)}"
res=qa_model(input_text)
print(res[0])
question="width of arc?"
input_text = f"question: {question} context: {' '.join(context)}"
res=qa_model(input_text)
print(res[0])
```
