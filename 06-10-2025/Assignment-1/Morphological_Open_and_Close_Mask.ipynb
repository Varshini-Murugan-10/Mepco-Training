# 2_Open_and_Close_Mask

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

mask = cv2.imread("/content/003.png", cv2.IMREAD_GRAYSCALE)

_, binary = cv2.threshold(mask, 127, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

kernel = np.ones((16, 16), np.uint8)
opened_mask = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
closed_mask = cv2.morphologyEx(opened_mask, cv2.MORPH_CLOSE, kernel)

cv2_imshow(mask)
cv2_imshow(opened_mask)
cv2_imshow(closed_mask)
```
