```python
# Otsu and Adaptive Thresholding
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("/content/000.png", cv2.IMREAD_GRAYSCALE)

ret, otsu_mask = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

mean_mask = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 15)
gaussian_mask = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 10)

titles = ['Original Image', "Otsu's Threshold", 'Adaptive Mean', 'Adaptive Gaussian']
images = [img, otsu_mask, mean_mask, gaussian_mask]

plt.figure(figsize=(10, 8))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap='gray')
    plt.axis('off')
plt.show()
```

```python
# SIFT and ORB Feature Detection
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("/content/000.png", cv2.IMREAD_GRAYSCALE)

sift = cv2.SIFT_create()
keypoints_sift, descriptors_sift = sift.detectAndCompute(img, None)
sift_img = cv2.drawKeypoints(img, keypoints_sift, None, flags=cv2.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)

orb = cv2.ORB_create()
keypoints_orb, descriptors_orb = orb.detectAndCompute(img, None)
orb_img = cv2.drawKeypoints(img, keypoints_orb, None, flags=cv2.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('SIFT Keypoints')
plt.imshow(sift_img)
plt.subplot(1, 2, 2)
plt.title('ORB Keypoints')
plt.imshow(orb_img)
plt.show()
```
